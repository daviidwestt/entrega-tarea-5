1) Implementar un juego, que consista en apilar números enteros del 1 al 20, de forma aleatoria, para lo cual debe usarse una estructura de Pila. 
Luego, el usuario debe elegir un número de veces en que se va a quitar elementos de la pila, los cuales, sumados entre sí, no deben superar el valor de 50.
El usuario pierde si la suma supera ese valor. Si no lo supera, gana, pero su calificación será 10 menos el número elementos que falten quitar para todavía no superar 50.
El programa debe informar si perdió, y si ganó, con qué calificación lo hizo.

Consideraciones:<br>
a. Se puede usar la función input() para obtener una entrada de teclado.<br>
b. Se puede usar la el modulo random para obtener valores aleatorios.

```python
import random

class Pila:
    def __init__(self):
        self.items = []

    def esta_vacia(self):
        return len(self.items) == 0

    def apilar(self, elemento):
        self.items.append(elemento)

    def desapilar(self):
        return self.items.pop()

def generar_pila_aleatoria():
    numeros = list(range(1, 21))
    random.shuffle(numeros)
    pila = Pila()
    for numero in numeros:
        pila.apilar(numero)
    return pila

def jugar():
    pila = generar_pila_aleatoria()
    suma_actual = 0
    intentos_restantes = int(input("Ingrese el número de veces que desea quitar elementos de la pila: "))

    while not pila.esta_vacia() and suma_actual <= 50 and intentos_restantes > 0:
        elemento = pila.desapilar()
        print(f"Elemento desapilado: {elemento}")
        suma_actual += elemento
        intentos_restantes -= 1

    if suma_actual > 50:
        print("¡Perdiste! La suma superó 50.")
    else:
        calificacion = 10 - intentos_restantes
        print(f"¡Ganaste! Calificación: {calificacion}")

if __name__ == "__main__":
    jugar()

```

2) Implementar un juego donde constas de 2 jarras, de capacidad 5 y 3 litros respectivamente, y debes colocar 4 litros en la jarra de 5L.
Las opciones posibles son:
* Llenar la jarra de 3 litros
* Llenar la jarra de 5 litros
* Vaciar la jarra de 3 litros
* Vaciar la jarra de 5 litros
* Verter el contenido de la jarra de 3 litros en la de 5 litros
* Verter el contenido de la jarra de 5 litros en la de 3 litros

```python
class Jarras:
    def __init__(self, capacidad_jarra_5L, capacidad_jarra_3L):
        self.jarra_5L = 0
        self.jarra_3L = 0
        self.capacidad_jarra_5L = capacidad_jarra_5L
        self.capacidad_jarra_3L = capacidad_jarra_3L

    def llenar_jarra_3L(self):
        self.jarra_3L = self.capacidad_jarra_3L

    def llenar_jarra_5L(self):
        self.jarra_5L = self.capacidad_jarra_5L

    def vaciar_jarra_3L(self):
        self.jarra_3L = 0

    def vaciar_jarra_5L(self):
        self.jarra_5L = 0

    def verter_3L_en_5L(self):
        espacio_disponible = self.capacidad_jarra_5L - self.jarra_5L
        if self.jarra_3L <= espacio_disponible:
            self.jarra_5L += self.jarra_3L
            self.jarra_3L = 0
        else:
            self.jarra_3L -= espacio_disponible
            self.jarra_5L = self.capacidad_jarra_5L

    def verter_5L_en_3L(self):
        espacio_disponible = self.capacidad_jarra_3L - self.jarra_3L
        if self.jarra_5L <= espacio_disponible:
            self.jarra_3L += self.jarra_5L
            self.jarra_5L = 0
        else:
            self.jarra_5L -= espacio_disponible
            self.jarra_3L = self.capacidad_jarra_3L

    def estado_actual(self):
        return f"Jarra de 5L: {self.jarra_5L}L, Jarra de 3L: {self.jarra_3L}L"


def jugar():
    jarras = Jarras(capacidad_jarra_5L=5, capacidad_jarra_3L=3)
    objetivo = 4

    while jarras.jarra_5L != objetivo:
        print(jarras.estado_actual())
        print("\nOpciones:")
        print("1. Llenar la jarra de 3 litros")
        print("2. Llenar la jarra de 5 litros")
        print("3. Vaciar la jarra de 3 litros")
        print("4. Vaciar la jarra de 5 litros")
        print("5. Verter el contenido de la jarra de 3 litros en la de 5 litros")
        print("6. Verter el contenido de la jarra de 5 litros en la de 3 litros")

        opcion = int(input("Seleccione una opción (1-6): "))

        if opcion == 1:
            jarras.llenar_jarra_3L()
        elif opcion == 2:
            jarras.llenar_jarra_5L()
        elif opcion == 3:
            jarras.vaciar_jarra_3L()
        elif opcion == 4:
            jarras.vaciar_jarra_5L()
        elif opcion == 5:
            jarras.verter_3L_en_5L()
        elif opcion == 6:
            jarras.verter_5L_en_3L()
        else:
            print("Opción no válida. Intente nuevamente.")

    print("¡Felicidades! Lograste colocar 4 litros en la jarra de 5 litros.")

if __name__ == "__main__":
    jugar()
```
